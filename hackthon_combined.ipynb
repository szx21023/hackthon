```python
from google.colab import drive
drive.mount('/content/drive')

import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def combined_image(path, image_name):
  img_0 = [cv2.imread(path + f'{image_name}_0_{j}.png') for j in range(8)]
  img_1 = [cv2.imread(path + f'{image_name}_1_{j}.png') for j in range(8)]
  img_2 = [cv2.imread(path + f'{image_name}_2_{j}.png') for j in range(8)]
  img_3 = [cv2.imread(path + f'{image_name}_3_{j}.png') for j in range(8)]
  img_4 = [cv2.imread(path + f'{image_name}_4_{j}.png') for j in range(8)]
  img_5 = [cv2.imread(path + f'{image_name}_5_{j}.png') for j in range(8)]
  img_6 = [cv2.imread(path + f'{image_name}_6_{j}.png') for j in range(8)]
  img_7 = [cv2.imread(path + f'{image_name}_7_{j}.png') for j in range(8)]
  vis_0 = np.concatenate(img_0, axis=1)
  vis_1 = np.concatenate(img_1, axis=1)
  vis_2 = np.concatenate(img_2, axis=1)
  vis_3 = np.concatenate(img_3, axis=1)
  vis_4 = np.concatenate(img_4, axis=1)
  vis_5 = np.concatenate(img_5, axis=1)
  vis_6 = np.concatenate(img_6, axis=1)
  vis_7 = np.concatenate(img_7, axis=1)
  vis = np.concatenate([vis_0, vis_1, vis_2, vis_3, vis_4, vis_5, vis_6, vis_7], axis=0)
  return vis
```

```python
image_name = 'output_15'
path = '/content/drive/MyDrive/side_project/satellite_segmentation/hackthon/image/'
vis = combined_image(path, image_name)
path = '/content/drive/MyDrive/side_project/satellite_segmentation/hackthon/label/'
vis_2 = combined_image(path, image_name)
path = '/content/drive/MyDrive/side_project/satellite_segmentation/hackthon/predict/'
vis_3 = combined_image(path, image_name)
# cv2.imwrite('out.png', vis)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 20))
ax1.imshow(vis)
ax2.imshow(vis_2*200)
ax3.imshow(vis_3*200)
```
