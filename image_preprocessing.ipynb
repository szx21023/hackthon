```python
import numpy as np
import glob
from PIL import Image
```

```python
def divide_image(folder, image_id, saved_folder):
    img = Image.open(folder + image_id + '.png')
    array = np.array(img.resize((2000, 2000)))

    for i in range(8):
        for j in range(8):
            img = Image.fromarray(array[250*i: 250*i+250, 250*j: 250*j+250, :3])
            path = saved_folder + f'{image_id}_{i}_{j}.png'
            img.save(path)
```

```python
image_path = f'/home/herry/hackthon/data/pic_atlas0519/'
label_path = f'/home/herry/hackthon/data/label/'

folder = f'/home/herry/hackthon/data/pic_atlas0519/'
saved_folder = '/home/herry/hackthon/data/preprocessing_image/image/'
files = glob.glob(folder + '*')

for i, image_path in enumerate(files):
    image_id = image_path.replace(folder, '').replace('.png', '')
    divide_image(folder, image_id, saved_folder)
    print((i + 1) / len(files))
```
